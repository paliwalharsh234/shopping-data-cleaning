```python
df = pd.read_csv("shirts.csv")

```

```python
df["discount"] = df["discount"].fillna(df["discount"].mean())
```

```python
df["seller_name"] = df["seller_name"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df[["discount","sizes"]]
```

```python
filtered_data = df[df["discount"] > 60.0]
filtered_data.head()
```

```python
df = df.drop_duplicates()
```

```python
df["price_category"] = df["initial_price"].apply(lambda x: "Expensive" if x > 2000.0 else "Affordable")
```

```python
df.head()
```

```python
df.to_csv("cleaned_shopping_data.csv", index=False)
```
